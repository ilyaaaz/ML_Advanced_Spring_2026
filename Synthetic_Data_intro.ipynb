# Synthetic data

In order to teach a model a task, we need ample quantities of training data
- features + labels: Supervised Learning
- features only: Self-Supervised/Unsupervised Learning

Without data: there is no training.

But there are impediments to gathering training examples.

## Scarcity

- Data may not be public
    - copyright/license issues
    - paywall
    - proprietary: code repo
    - Sensitive: email, text messages
- The task is brand new and no "real-world" data is available
- Data is available, but in insufficient quantity.

The larger/more complex the model: the more training data is needed
- prevent under-fitting
- to full describe a more complex functions with empirical examples, more examples are needed

The scarcity is most apparent when training large LLMs.

Scaling laws for LLMs state that 
- for a fixed compute budget $C$
- to achieve the minimum training Loss
- the amount of training data $D$ is linear in the number of model parameters $N$

$$
D = 20 * N
$$

For a 1 trillion parameter model, we need 20 Trillion tokens !
- there are only about 30 trillion human-generated tokens
- reusing tokens is possible, but at a loss of efficiency

## Privacy

There are many cases where use of data is constrained by legal or ethical concerns
- Medical record privacy (HIPAA)
- Local regulations (GDPR)


## Robustness

There are cases where 
- there is a lot of "high probability" examples are available
- but the "low probability" examples are the ones needed for training

For example: to predict a market crash, I need many training examples of extreme moves in markets.

Our Fundamental Law of Machine Learning hypothesizes that
- real examples of a task come from a distribution $\pdata$
- our training data is an empirical sample from this distribution

But empirical data is necessarily going to have limited examples of low probability events.
- for a Risk Manager: those may be the most relevant

# Solution: Create Synthetic examples for training

The solution to these issues is to
- create models
- whose output are Synthetic examples
    - ideally: indistinguishable from "natural" examples
    

The synthetic data generation model will vary by type of data
- text: LLM's
- numeric
    - regression
    - Auto-Encoders
- image
    - GANs
