# Minimum tasks

## Model 1: The Pre-trained model

Instantiate the pre-trained model together with an appropriate classifier head.

We will refer to this a "Model PT"

## Model 2: Pre-trained model + fine-tuned Classifier "head"

Starting with the previous model "Model PT"
- freeze all the weights *except* for the weights in the Classifier head
    - the head may be more than one layer
- fine-tune just the weights of the Classifier head

We will refer to the trained version of this as "Model PT + FT Classifier"

## Model 3: Fine-tuned model with your own classifier head

The purpose of this part is for you to demonstrate competency in the Functional architecture.

Starting with the model "Model PT"
- *remove* the existing Classifier head
- *replace* it with a Classifier head of your own design
- freeze all the weights *except* for the newly added Classifier head
- fine-tune just the weights in the newly added Classifier head

It is *required* for you
- to create a Classifier head that is different than the original
- *explain* why it is different and what inspired your design choices
- *explain* the technical challenge of grafting the head onto the pre-existing body

We will refer to the trained version of this as "Model PT + Own FT Classifier"

## Model 4: Pre-trained model + fine-tuned Classifier head + fine-tuned body

Starting with the model "Model PT + FT Classifier"
- freeze the weights of the Classifier head
- fine-tune all other weights

We will refer to the trained versions of this as "Model PT + FT Classifier + FT Base"

## Model 5: Pre-trained model + fine-tuned Classifier head + PEFT base

This is a variation of the task that created "Model PT + FT Classifier + FT Base".

Starting with the model "Model PT + FT Classifier"
- freeze the weights of the Classifier head
- fine-tune all other weights *using LoRA* with the following restrictions
    - you will fine-tune **only the weights** of the FFN blocks in every layer
        - this is where we hypothesize "world knowledge" acquired during training is stored
        - each FFN block has 2 `Dense` layers: `lin1`, `lin2`
            - apply LoRA **to each** of the 2 `Dense` layers
        - use rank 8 for the LoRA adapters
        
Additional requirements
- the objective is for you to demonstrate competence with the Functional architecture
    - you must *manually* create the adapters and insert them into the model
        - you **may not** use any tool that automatically does this (e.g. HuggingFace has PEFT tools)
- you should manually calculate the number of trainable weights in this configuration
    - report this number, along with your calculation
    - verify that your model matches this calculation
