# Synthetic Text Data: case study - Instruction Following task

An LLM is a "predict the next token" machine.
-  trained on the Language Modeling task 

A human wanting to know about the Black-Scholes equation
- would need to formulate the request following the "predict the next" paradigm (Universal API)

        The Black-Scholes equation states ...

We would find this frustrating ! 

We transform the LLM into a "helpful assistant" by Fine-Tuning it in the *Instruction Following* task
- given an instruction and an instance of a task
- solve the task by following the instructions

For example
- "Summarize the following document"
- "Rank the following possible answers to a given question"

Assistant behavior allows the human to pose the more natural form of the Black Scholes query

        What does the Black-Scholes equation state ?

where the Assistant follows the implicit instruction to answer questions in a helpful, honest and harmless manner
    

But where do the training examples
- for the *Instruction Following* task

come from ?

There is little (if any) "natural" training data for this task.

We use this Fine Tuning task's data requirements as 
a *case study*
for the topic of *Synthetic Text Data*

# Backtranslation

**Reference**

- [Self-Alignment with Instruction Backtranslation](https://arxiv.org/pdf/2308.06259.pdf)

<br>
We now present an alternate method for generating the Instruction part of 
$$
\langle \text{Instruction} \rangle
$$
of an Instruction Following triple
$$
\langle \text{Instruction}, \text{Context}, \text{Response} \rangle
$$


We start off with an LLM $M_{xy}$ acting as a "predict the next token" machine
- trained on $\langle \x, \y \rangle$ pairs
- given input $\x$ (Instruction)
- generates output $\y$ (Response)

which generates $\langle \text{Instruction}, \text{Response} \rangle$ pairs

We then create an "inverse" LLM $M_{yx}$
- trained on $\langle \y, \x \rangle$ pairs
- given input $\y$ (Response)
- generates output $\x$ (Instruction)

This is called *Back-translation*

Using $M_{yx}$
- we can treat a block of text $\y$ as the Response
- and generate the  $\langle \text{Instruction}, \text{Context} \rangle$ pair $\x$
    - which is the Instruction/Context for which $y$ is an appropriate response

The advantage of this technique
- virtually any block of text can play the role of $\y$
- giving an abundant source to generate $\langle \x, \y \rangle$ pairs.

This is similar in spirit to the Language Modeling task
- any block of text can become the 
$$\x = \x_{(0)}, \ldots \x_{(\tt-1)}$$
with known target $\x_\tp$

# Preference Data: RLHF

When given a prompt whose response is "non-verifiable"
- there are many possible acceptable responses
- it is a matter of *preference* as to which is the "best"

In addition to the Instruction Following task, creating models to rank outputs is important
- steer a model to have desirable characteristics
    - honest
    - helpful
    - harmless
    

In order to Fine Tune an LLM to follow preferences
- we need to create a *Preference Dataset*
    - prompt
    - two candidate responses
    - a label indicating the preferred response
    
    

For example

**Prompt:** "How do I change a tire?"
- **Reply A:** An accurate step-by-step answer.
- **Reply B:** A brief, incomplete answer.

Both replies are "correct" but the first is subjectively better.

An example of *Preference Data* is a triple

$$
(x, y^+, y^-)
$$
- input $x$
- the preferred output $y^+$
- the non-preferred output $y^-$


Generating synthetic Preference Data examples takes a similar path to the 
generation of synthetic Instruction Following data
- start with human labeled data (Step 1 in the diagram below)
- use it to train a Reward/Preference model to rank responses (Step 2 in the digram)
- bootstrap/self-improvement to generate more preference examples

We will learn more about the uses for Preference Data in our module on Post-Training an LLM.
