# The Model Context Protocol (MCP)
## The "Nervous System" of Agentic Architecture

The **Model Context Protocol (MCP)** is an open-standard
- JSON-RPC based communication layer

It acts as  **standardized interface** 
- that allows a Client (the Agent/Brain)
- to interact with diverse Environment Servers (Databases, File Systems, APIs)

without needing custom "glue code" for every tool.

**Key Technical Pillars:**
* **Decoupling:** 
    - The Brain doesn't need to know *how* a database works
    - it only needs to know how to send an MCP `call_tool` request.
* **Server-Side Logic:**
    - MCP allows logic (and intermediate steps) to live on the server.
        - Instead of the Brain fetching 10,000 lines of logs
        - the MCP Server runs a local filter
    - and returnsa "Golden Nugget"
        - a 10-line summary
    - instead of a "Firehose" of intermediate results
* **Standardized Discovery:** 
    - The Client can query an MCP Server to "discover" its capabilities dynamically 
         -`list_tools`, `list_resources`

# Skills (`skills.md`)

## The "Muscle Memory" of Agentic Reasoning

While MCP is about **access**
- a **Skill** is about **competency**. 

Based on the Anthropic "Skills" pattern, a skill is
- a high-level instruction set
    — often stored in a `skills.md` file
- that describes to the LLM 
    - exactly how to carry out a specialized, multi-step task.

**Key Components of a Skill:**
* **Heuristics:**
    - Expert "rules of thumb" 
        - e.g., "Always check the `package.json` before refactoring a dependency".
* **Workflow Scaffolding:** 
    - A step-by-step guide on how to chain tools together to achieve a complex goal.
* **Constraints:** 
    - Guardrails that prevent the model from taking "naive" actions 
        - e.g., "Do not run `rm -rf` without first listing the directory contents".

# Comparison: Protocol vs. Pedagogy

The following table contrasts the two approaches to help determine where to place logic in your Harness Engineering.

| Feature | MCP (The Protocol) | Skills (The Pedagogy) |
| :--- | :--- | :--- |
| **Primary Role** | **Transport & Labor:** Moving and filtering data. | **Strategy & Wisdom:** Deciding how to act. |
| **Where it lives** | In the **Harness/Server** (Python/TypeScript code). | In the **Prompt/Context** (Markdown/Text). |
| **Token Cost** | **Low:** Offloads processing to external compute. | **High:** Consumes context window with instructions. |
| **Security** | **Hard Enforcement:** The server can deny requests. | **Soft Enforcement:** The model is "asked" to be safe. |
| **Best For** | Complex data retrieval, RAG, and API interaction. | Domain-specific workflows and expert decision-making. |


**MCP gives the agent Ability; Skills give the agent Wisdom.**
- You use a **Skill** to teach the Brain *what* to ask for
- You use **MCP** to ensure the answer is delivered as a "Golden Nugget" rather than a "Firehose of Noise."
